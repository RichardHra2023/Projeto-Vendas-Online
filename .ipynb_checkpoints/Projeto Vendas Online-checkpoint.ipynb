# Projeto Vendas Online

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dicionário de dados

Order ID: Unique identifier for each sales order.

Date: Date of the sales transaction.

Category: Broad category of the product sold (e.g., Electronics, Home Appliances, Clothing, Books, Beauty Products, Sports).

Product Name: Specific name or model of the product sold.

Units Sold: Number of units of the product sold in the transaction.

Unit Price: Price of one unit of the product.

Total Revenue: Total revenue generated from the sales transaction (Quantity * Unit Price).

Region: Geographic region where the transaction occurred (e.g., North America, Europe, Asia).

Payment Method: Method used for payment (e.g., Credit Card, PayPal, Debit Card).

### Fonte da base de dados:

https://www.kaggle.com/datasets/shreyanshverma27/online-sales-dataset-popular-marketplace-data/data

### Vamos analisar a distribuição dos preços dos produtos:

```python
sns.boxplot(data = df, x = None, y = 'Unit Price' )
```

```python
media_produto = df['Unit Price'].mean()
mediana_produto = df['Unit Price'].median()
desv_padrao = df['Unit Price'].std()

print('Os Produtos vendidos tem um preço médio de $ {}, com mediana de $ {} e desvio padrão de $ {}'
      .format(round(media_produto ,2), round(media_produto ,2), round(desv_padrao, 2)))
```

Podemos observar no grafico acima que existem muitos outliers no boxplot. Isso indica que existem varios produtos que tem um preço muito acima da mediana. Iremos ver em breve quais são esses produtos, mas podemos perceber que os produtos eletronicos e eletrodomesticos que são os mais caros dentre todas as categorias.

## Analisando as vendas durante o ano:

### Vamos analisar a quantidade de vendas e a receita total durante o ano.

```python
# Criando o gráfico de linhas para análise da quantidade de Vendas durante o ano.
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Units Sold'], color='green')

# Configurando os rótulos e o título
plt.xlabel('')
plt.ylabel('Total de Unidades Vendidas', size = 18)
plt.title('Quantidade de Vendas durante o Ano', size = 25)
plt.xticks(rotation=45)
plt.ylim(0, 12)
plt.tight_layout()
plt.show()
```

```python
# Criando o gráfico de linhas para analise da Receita total durante o ano.
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Total Revenue'], color='blue')

# Configurando os rótulos e o título
plt.xlabel('')
plt.ylabel('Receita', size = 18)
plt.title('Receita Total obtida durante o Ano', size = 25)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Analisando a quantidade de produtos vendidas por categoria

```python
total_vendido_por_categoria = df.groupby('Product Category')['Units Sold'].sum().reset_index()
```

```python
total_vendido_por_categoria
```

```python
# Criando o gráfico de barras para analise.
plt.figure(figsize=(10, 6))
bars = plt.bar(total_vendido_por_categoria['Product Category'], total_vendido_por_categoria['Units Sold'], color='blue')

# Adicionando rótulos de dados nas barras para facilitar a leitura dos valores
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')

# Configurando os rótulos e o título
plt.xlabel('Categoria do Produto')
plt.ylabel('Total de Unidades Vendidas')
plt.title('Total de Unidades Vendidas por Categoria de Produto')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Analisando a quantidade de vendida por método de pagamento

```python
total_vendido_met_pagamento = df.groupby('Payment Method')['Units Sold'].sum().reset_index()
```

```python
total_vendido_met_pagamento
```

```python
# Criando o gráfico de barras para a analise.
plt.figure(figsize=(10, 6))
bars = plt.bar(total_vendido_met_pagamento['Payment Method'], total_vendido_met_pagamento['Units Sold'], color='green')

# Adicionando rótulos de dados nas barras para facilitar a leitura dos valores
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')

# Configurando os rótulos e o título
plt.xlabel('Método de Pagamento')
plt.ylabel('Total de Unidades Vendidas')
plt.title('Total de Unidades Vendidas por Método de pagamento')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Analisando a quantidade vendida e a Receita por Região

```python
total_qtd_vendido_regiao = df.groupby('Region')['Units Sold'].sum().reset_index()
total_vendido_regiao = df.groupby('Region')['Total Revenue'].sum().reset_index()
```

```python
total_qtd_vendido_regiao
```

```python
total_vendido_regiao
```

```python
# Criando o gráfico de barras para analise.
plt.figure(figsize=(10, 6))
bars = plt.bar(total_qtd_vendido_regiao['Region'], total_qtd_vendido_regiao['Units Sold'], color='skyblue')

# Adicionando rótulos de dados nas barras para facilitar a leitura dos valores
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')

# Configurando os rótulos e o título
plt.xlabel('Região')
plt.ylabel('Total de Unidades Vendidas')
plt.title('Total de Unidades Vendidas por Região')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Criando o gráfico de barras para analise.
plt.figure(figsize=(10, 6))
bars = plt.bar(total_vendido_regiao['Region'], total_vendido_regiao['Total Revenue'], color='orange')

# Adicionando rótulos de dados nas barras para facilitar a leitura dos valores
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')

# Configurando os rótulos e o título
plt.xlabel('Região')
plt.ylabel('Receita da Venda')
plt.title('Valor Total Vendido por Região')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Criando a figura com 2 subplots lado a lado para comparação.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

# Primeiro gráfico - Total de Unidades Vendidas por Região
axes[0].bar(total_qtd_vendido_regiao['Region'], total_qtd_vendido_regiao['Units Sold'], color='skyblue')
for bar in axes[0].containers[0]:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')
axes[0].set_xlabel('Região')
axes[0].set_ylabel('Total de Unidades Vendidas')
axes[0].set_title('Total de Unidades Vendidas por Região')
axes[0].tick_params(axis='x', rotation=45)

# Segundo gráfico - Valor Total Vendido por Região
axes[1].bar(total_vendido_regiao['Region'], total_vendido_regiao['Total Revenue'], color='orange')
for bar in axes[1].containers[0]:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')
axes[1].set_xlabel('Região')
axes[1].set_ylabel('Valor Vendido')
axes[1].set_title('Valor Total Vendido por Região')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Podemos ver no gráfico acima que a Região da Europa é a que menos comprou produtos e foi a que menos teve lucro. Já a Região da Ásia que apesar de ter o maior volume em quantidade de produtos vendidos, teve um lucro aproximado com a da Europa. Já a América do Norte foi a região onde mais obteve lucro.

### Analisando o lucro por categoria de produto.

```python
lucro_categoria = df.groupby('Product Category')['Total Revenue'].sum().reset_index()
```

```python
lucro_categoria
```

```python
# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(lucro_categoria['Product Category'], lucro_categoria['Total Revenue'], color='yellow')

# Adicionando rótulos de dados nas barras para facilitar a leitura dos valores
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

# Configurando os rótulos e o título
plt.xlabel('Categoria do Produto')
plt.ylabel('Receita')
plt.ylim(0, 40000)
plt.title('Receita total por Categoria de Produto')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Vamos analisar o lucro por tipo de método de pagamento.

```python
lucro_metodo_pagamento_regiao = df.groupby(['Region', 'Payment Method'])['Total Revenue'].sum().reset_index()
```

```python
lucro_metodo_pagamento_regiao
```

```python
# criando o grafico de barras usando seaborn

plt.figure(figsize = (10,8))
ax = sns.barplot(x = 'Region', y = 'Total Revenue', hue = 'Payment Method', data = lucro_metodo_pagamento_regiao, palette = 'viridis')

# Adicionando rótulos de dados nas barras para facilitar a leitura dos valores
for container in ax.containers:
    ax.bar_label(container, fmt = '%.2f', padding = 3)
    
# Configurando os rótulos e o título
plt.xlabel('Região')
plt.ylabel('Receita')
plt.title('Receita Total por Região e Tipo de Pagamento', size = 20)
plt.xticks(rotation=45)
plt.ylim(0, 40000)
plt.legend(title='Categoria do Produto')
plt.tight_layout()
plt.show()
```

Podemos observar no grafico acima que a região da américa no norte foi a que obteve mais lucro e foi a que usou um unico método de pagamento que foi cartão de crédito. A região da Europa tambem utilizou um unico método de pagamento que foi o PayPal, já a região da Ásia utilizou 2 metodos de pagamento que foram os cartões de crédito e débito e ficando em segundo lugar no total de receita.

### Vamos analisar o lucro por Região e Categoria de Produto

```python
regiao_categoria_df = df.groupby(['Region', 'Product Category'])['Total Revenue'].sum().reset_index()
```

```python
# Criando o gráfico de barras usando seaborn
plt.figure(figsize=(10, 8))
ax = sns.barplot(x='Region', y='Total Revenue', hue='Product Category', data=regiao_categoria_df, palette='viridis')

# Adicionando rótulos de dados nas barras para facilitar a leitura dos valores
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3, rotation=90)

# Configurando os rótulos e o título
plt.xlabel('Região')
plt.ylabel('Receita')
plt.title('Receita Total por Região e Categoria de Produto', size = 20)
plt.xticks(rotation=45)
plt.ylim(0, 40000)
plt.legend(title='Categoria do Produto')
plt.tight_layout()
plt.show()
```

Podemos ver que o gráfico acima responde o porque da região da América do Norte ter tido mais lucro que o restante. Vimos que em sua maioria, as vendas feitas nessa região foi de produtos da categoria de eletronicos, uma vez que essa categoria de produto é a mais cara dentre as outras que existem no portfolio da loja. Apesar da categoria de Eletrodomesticos tambem terem um valor alto, as vendas foram menores.
